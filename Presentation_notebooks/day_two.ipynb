```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
from GetData.read_data import get_stock_data
```

```python
Apple = get_stock_data()
print(type(Apple))
```

```python
# Apple.Open gives you the y values.
Apple_index = pd.to_numeric(Apple.index)

plt.figure(0, figsize=(12, 5))
plt.title("Stock Price vs Time", fontsize="x-large")
plt.plot(Apple.index, Apple.Open)
plt.legend(loc=2)
plt.show()
```

```python
y = (Apple.Open > np.mean(Apple.Open)).astype(int)

plt.figure(1, figsize=(12, 5))
plt.title("Logistic Plot of Stock Greater Than the Mean Price", fontsize="x-large")
plt.scatter(Apple.index, y)
plt.legend(loc=2)
plt.show()
```

```python
logistic_reg = LogisticRegression(solver='lbfgs')
logistic_reg.fit(Apple.Open, y)
```
